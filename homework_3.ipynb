### 2.Verimizde yıl (year) sütunu olduğunu farketmişsinizdir. Şimdilik yıl verisini unutun ve her bir gözlemin aynı yıl içerisinde yapıldığını farz edin. Her bir değişken için eksik değerleri nasıl doldurabileceğinizi düşünün. Eksik değerleri bir değerle doldurmak hangi değişkenler için anlamlı, hangileri için anlamsızdır?

```python
## eksik veriler tüm veri setinin önemli bölümünü kapsadığı için onları silmek veri kaybına neden olur bu yüzden ortalama ile doldurulabilir.
```

```python
state1 = state.copy()
```

```python
for sutun in state1.columns[3:,]:                             
                                                              
    state1[sutun].fillna(state1[sutun].mean(),inplace=True)
```

```python
state1
```

```python
state1.isnull().sum()*100/state1.shape[0]
```

### 3.Şimdi zaman faktörünü dikkate alma zamanı! 2. sorudaki cevabınızı tekrar gözden geçirin ve eksik verileri o yıl içerisinde gözlemlenen değerlere dayanarak doldurun. Örneğin, bir değeri ortalama değer ile doldurmak isterseniz, o yılın ortalamasını hesaplayın.

```python
state2 = state.copy()
```

```python
yil_ort = state2.groupby("YEAR").mean()
yil_ort
```

```python
for sutun in state2.columns[3:,]:
    
    for i in state2["YEAR"].unique():
        
        state2[sutun][(state2["YEAR"]==i) & (state2[sutun].isnull())] = yil_ort[sutun][i]
        
```

```python
state2
```

```python
state2.isnull().sum()*100/state2.shape[0]
```

### 4.Bu sefer, eksik değerleri enterpolasyon yaparak doldurun.

```python
### eksik verilerin enterpolasyon ile doldurulması
```

```python
state3 = state.copy()
```

```python
for eyalet in state3["STATE"].unique():
    state3[state3["STATE"] == eyalet] = state3[state3["STATE"] == eyalet].interpolate(limit_direction = "both")
        
state3.head()   
```

```python
state3.isnull().sum()*100/state3.shape[0]
```
