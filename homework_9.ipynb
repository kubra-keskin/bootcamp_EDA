```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
yazı_tura = []
yazı_tura_ort = []
for atış_sayısı in range(1,1000):
    atış = np.random.binomial(1,.5)
    yazı_tura.append(atış)
    yazı_tura_ort.append(np.mean(yazı_tura))
    
print(yazı_tura[:10])
print(yazı_tura_ort[:50])
```

```python

baslik_font = {'family': 'arial', 'color': 'darkred','weight': 'bold','size': 13 }
eksen_font  = {'family': 'arial', 'color': 'darkblue','weight': 'bold','size': 10 }

plt.hlines(0.5, 0, 1000, colors= 'r' )
plt.plot(yazı_tura_ort)
plt.ylim((0,1))
plt.title("Örneklem Sayısı = 1000" ,fontdict = baslik_font)
plt.xlabel("Örneklem Sayısı", fontdict = eksen_font)
plt.ylabel("Örneklem Ortalaması", fontdict = eksen_font)
```
